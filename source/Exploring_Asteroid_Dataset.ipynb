# Data Preprocessing

Dropping columns with non-numerical values to ease plotting.

Additionally, changing Y and N values in 'pha' feature to 0 and 1 for numerical values.

Dropping all rows with NaN diameter values, along with converting non-NaN diameter values to float

```python
sel_data = data.drop(columns=["name", "spec_B", "spec_T", "neo", "G", "class", "extent", "IR", "condition_code", "n_obs_used", "data_arc"])
sel_data['pha'] = sel_data['pha'].map({'Y': 1, 'N': 0})

#Dropping rows with missing diameter values
sel_data = sel_data.dropna(subset=['diameter'])
```

```python
#Converting diameter values from object to float
sel_data['diameter'] = sel_data['diameter'].str.replace('km', '').astype(float)
```

```python
#Remove outliers

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sel_data_norm = pd.DataFrame(scaler.fit_transform(sel_data), columns = sel_data.columns)

display(sel_data)
display(sel_data_norm )
```

# Model 1: Linear Regression

```python
sel_data1 = sel_data_norm[['q', 'moid', 'H', 'diameter']]
sel_data1 = sel_data1.dropna()

sel_data1
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(sel_data1, test_size=0.2, random_state=42)


from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from scipy import stats


X_train = train_df[['H']]
# X_train = train_df[['H']]
y_train = train_df['diameter']


# X_test = test_df[['H']]
X_test = test_df[['H']]
y_test = test_df['diameter']

model = linear_model.LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print(f"Linear- MSE: {mse}")
```

```python
plt.scatter(X_test['H'], y_test, color="black")
plt.plot(X_test['H'], y_pred, color="blue", linewidth=3)

plt.show()
```

## Model 2 - Polynomial Regression between H and Diameter - Degree 11

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


degree =6
poly = PolynomialFeatures(degree)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

model = LinearRegression()
model.fit(X_train_poly, y_train)

y_train_pred = model.predict(X_train_poly)
y_test_pred = model.predict(X_test_poly)

topplotx = np.linspace(X_train.min(), X_train.max(), 100)
topploty = model.predict(poly.transform(topplotx.reshape(-1, 1)))

mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)


plt.scatter(X_test, y_test, color="black")
plt.plot(topplotx, topploty)

plt.xticks(())
plt.yticks(())

plt.show()

print(f"Degree {degree} - Polynomial-Training MSE: {mse_train}, Polynomial-Testing MSE: {mse_test}")


```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

def train_and_eval(X_train, y_train, X_test, y_test, degree):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    poly = PolynomialFeatures(degree)
    X_train_poly = poly.fit_transform(X_train_scaled)
    X_test_poly = poly.transform(X_test_scaled)

    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)

    mse_train = mean_squared_error(y_train, y_train_pred)
    mse_test = mean_squared_error(y_test, y_test_pred)

    print(f"Degree {degree} Train MSE: {mse_train:.6f}, Test MSE: {mse_test:.6f}")
    return model.coef_, mse_train, mse_test

train_df, test_df = train_test_split(sel_data1, test_size=0.2, random_state=42)

X_train = train_df[['q','H']]
y_train = train_df['diameter']
X_test = test_df[['q','H']]
y_test = test_df['diameter']
model_coef = train_and_eval(X_train, y_train, X_test, y_test, 6)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(units = 16, activation = 'relu', input_dim = 2)) # here our input dimension is set to 9 via the input_dim parameter for our first hidden layer
classifier.add(Dense(units = 8, activation = 'relu'))
classifier.add(Dense(units = 6, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'relu'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'mean_squared_error')
```
