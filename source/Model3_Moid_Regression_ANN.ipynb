# Data Preprocessing

Dropping columns with non-numerical values to ease plotting.

Additionally, changing Y and N values in 'pha' feature to 0 and 1 for numerical values.

Dropping all rows with NaN diameter values, along with converting non-NaN diameter values to float

```python
sel_data = data.drop(columns=["name", "spec_B", "spec_T", "neo", "G", "extent", "IR", "condition_code", "n_obs_used", "data_arc"])
sel_data['pha'] = sel_data['pha'].map({'Y': 1, 'N': 0})
sel_data['class']=sel_data['class'].map({'MBA':0, 'OMB':1, 'MCA':2, 'AMO':3, 'IMB':4, 'TJN':5, 'CEN':6, 'APO':7, 'ATE':8,'AST':9, 'TNO':10, 'IEO':11, 'HYA':12, 'PAA':13})
display(sel_data["class"].unique())
#Dropping rows with missing diameter values
sel_data = sel_data.dropna(subset=['diameter'])
sel_data.dtypes
print(sel_data["class"].unique())
```

```python
#Converting diameter values from object to float
sel_data['diameter'] = sel_data['diameter'].str.replace('km', '').astype(float)
sel_data
```

```python
# I want to get rid of the very big values in a and diameter that might mess up the normalization of the data
sel_data = sel_data[sel_data['a'] < 10]
sel_data = sel_data[sel_data['diameter'] < 200]
```

```python
#Normalizing Data

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sel_data_norm = pd.DataFrame(scaler.fit_transform(sel_data), columns = sel_data.columns)

display(sel_data)
display(sel_data_norm)
```

## Neural Network for predicting ``moid``

```python
sel_data_neural = sel_data_norm.drop(columns = ['a','om','w','q','ad','pha','per_y','rot_per','GM','BV','UB','class','n','per','ma'])
```

```python
sel_data_neural.columns
sel_data_neural = sel_data_neural.dropna(subset=['albedo'])

```

```python
# !pip3 install scikeras
from sklearn.model_selection import RandomizedSearchCV
from scikeras.wrappers import KerasClassifier
from scikeras.wrappers import KerasRegressor
from keras.models import Sequential
from keras.layers import Dense
from scipy.stats import randint
import warnings
# Suppress specific user warnings
warnings.filterwarnings('ignore', message='``build_fn`` will be renamed to ``model`` in a future release')
warnings.filterwarnings('ignore', message='Do not pass an `input_shape`/`input_dim` argument to a layer.')
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(sel_data_neural, test_size=0.2, random_state=42)

X_train = train_df.drop(columns=['moid'])
y_train = train_df['moid']

X_test = test_df.drop(columns=['moid'])
y_test = test_df['moid']
```

Add in HP-tuning

```python
# set up HP-tuning
def create_model(units1=32, units2=16, units3=8, optimizer='adam'):
    model = Sequential()
    model.add(Dense(units=units1, activation='relu', input_dim=X_train.shape[1]))
    model.add(Dense(units=units2, activation='relu'))
    model.add(Dense(units=units3, activation='relu'))
    model.add(Dense(units=4, activation='relu'))
    model.add(Dense(units=2, activation='relu'))
    model.add(Dense(units=1, activation='relu'))
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

param_dist = {
    'units2': randint(8, 32),
    'units3': randint(4, 16),
    'batch_size': [10, 20],
    'epochs': [10, 50],
    'optimizer': ['adam','rmsprop']
}

# Suppress all warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

model = KerasRegressor(build_fn=create_model, verbose=0,units1=38,units2=16,units3=8)

random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=100, cv=3, random_state=42, n_jobs=-1)
random_search_result = random_search.fit(X_train, y_train)





```

```python
print("Best: %f using %s" % (random_search_result.best_score_, random_search_result.best_params_))
```

```python
classifier = Sequential()
classifier.add(Dense(units = 32, activation = 'relu', input_dim = X_train.shape[1]))
classifier.add(Dense(units = 22, activation = 'relu'))
classifier.add(Dense(units = 9, activation = 'relu'))
classifier.add(Dense(units = 4, activation = 'relu'))
classifier.add(Dense(units = 2, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'relu'))
```

```python
classifier.compile(optimizer = 'rmsprop', loss = 'mean_squared_error')
```

```python
classifier.summary()
```

```python
history = classifier.fit(train_df.astype(float).drop(columns = ['moid']), train_df.astype(float)['moid'], batch_size = 20, epochs = 50)
```

```python
yhat_test = classifier.predict(test_df.drop(columns = ['moid']))
```

```python
y_test = test_df['moid'].astype(float)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, yhat_test)
print("Neural-MSE",mse)

#Calculating training MSE
yhat_train = classifier.predict(train_df.drop(columns = ['moid']))
mse_train = mean_squared_error(train_df['moid'], yhat_train)
print("Training-MSE",mse_train)
```
